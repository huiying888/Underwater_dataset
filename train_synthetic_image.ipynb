```python
import sys
import os
import cv2
import numpy as np
import glob
from tqdm.notebook import tqdm
import torch
import torchvision

# --- CRITICAL ENVIRONMENT CHECK ---
PYTHON_EXECUTABLE = sys.executable

print(f"1. Using Python Executable: {PYTHON_EXECUTABLE}")
print(f"2. PyTorch Version: {torch.__version__}")
print(f"3. CUDA Available: {torch.cuda.is_available()}")

if not torch.cuda.is_available():
    print("\n--- WARNING: CUDA NOT FOUND. TRAINING WILL BE EXTREMELY SLOW. ---")
    
# --- CONFIGURATION ---
IMAGE_SIZE = (640, 640) 
PROJECT_ROOT = os.getcwd()
PIX2PIX_ROOT = os.path.join(PROJECT_ROOT, 'datasets', 'sonar_pix2pix-Copy1')
COMBINED_ROOT = os.path.join(PROJECT_ROOT, 'datasets', 'sonar_pix2pix_combined_v2')

# Domain definitions for the generator script
ORIGINAL_B_ROOT = os.path.join(PIX2PIX_ROOT, 'B')
TARGET_A_ROOT = os.path.join(PIX2PIX_ROOT, 'A')

# Create the required final structure
os.makedirs(os.path.join(TARGET_A_ROOT, 'train'), exist_ok=True)
os.makedirs(os.path.join(TARGET_A_ROOT, 'test'), exist_ok=True)
os.makedirs(os.path.join(COMBINED_ROOT), exist_ok=True)

print(f"4. Original Data (Domain B) Expected Path: {ORIGINAL_B_ROOT}")
print("   Please ensure your realistic sonar images are placed in B/train/ and B/test/")
```

```python
from tqdm import tqdm
def generate_canny_edge_mask(image_path):
    """
    Applies Canny edge detection, smoothing, and inversion to create the 
    geometric mask (Domain A) from the original photo (Domain B).
    """
    # 1. Load and Resize Image
    image = cv2.imread(image_path)
    if image is None:
        return None
    
    # Resize to the standard training dimension (CRITICAL for Pix2pix pairing)
    image = cv2.resize(image, IMAGE_SIZE, interpolation=cv2.INTER_AREA)

    # 2. Preprocessing
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # 3. Apply Canny Edge Detection (You can tune these thresholds)
    edges = cv2.Canny(gray_image, 100, 200)

    # 4. Apply Morphological Operations to Smooth/Connect Edges
    kernel = np.ones((3, 3), np.uint8)
    edges = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, kernel)

    # 5. Invert Colors (For a white line/feature on a black background)
    edges_inverted = cv2.bitwise_not(edges)

    # 6. Convert to 3-Channel RGB (Required by most GAN frameworks)
    mask_rgb = cv2.cvtColor(edges_inverted, cv2.COLOR_GRAY2BGR)
    
    return mask_rgb


def process_and_save_data(split_name):
    """Handles the processing and saving of masks for a given split (train/test)."""
    source_dir_B = os.path.join(ORIGINAL_B_ROOT, split_name)
    target_dir_A = os.path.join(TARGET_A_ROOT, split_name)

    if not os.path.exists(source_dir_B):
        print(f"--- WARNING: Source directory B/{split_name} not found. Skipping. ---")
        return

    os.makedirs(target_dir_A, exist_ok=True)
    
    image_paths = glob.glob(os.path.join(source_dir_B, '*.png'))
    image_paths.extend(glob.glob(os.path.join(source_dir_B, '*.jpg')))

    print(f"Found {len(image_paths)} images in B/{split_name}. Creating paired masks...")
    
    for path in tqdm(image_paths):
        mask = generate_canny_edge_mask(path)
        
        if mask is not None:
            filename = os.path.basename(path)
            output_path = os.path.join(target_dir_A, filename)
            cv2.imwrite(output_path, mask)

    print(f"Successfully created {len(image_paths)} masks for A/{split_name}.")

# --- EXECUTE MASK GENERATION ---
process_and_save_data('train')
process_and_save_data('test')

print("\nMask Generation Complete (Domain A is ready).")
```
